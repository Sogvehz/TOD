### JSON

1.1 Считайте файл `contributors_sample.json`. Воспользовавшись модулем `json`, преобразуйте содержимое файла в соответствующие объекты python. Выведите на экран информацию о первых 3 пользователях.

````python
import json
with open("data/contributors_sample.json", 'r') as f:
    k = json.load(f)
    print(k[:3])
    f.close()
k
````

1.2 Выведите уникальные почтовые домены, содержащиеся в почтовых адресах людей

````python
mails = {}
for i in k:
    mails.update({i["mail"][i["mail"].find("@") + 1:]: 0})
mails.keys()
````

1.3 Напишите функцию, которая по `username` ищет человека и выводит информацию о нем. Если пользователь с заданным `username` отсутствует, возбудите исключение `ValueError`

````python
def find_man(username):
    for i in k:
        if i["username"] == username:
            return i["name"]
    raise ValueError
find_man("uhebert")
````

1.4 Посчитайте, сколько мужчин и женщин присутсвует в этом наборе данных.

````python
man, woman = 0, 0
for i in k:
    if i["sex"] == "M":
        man += 1
    elif i["sex"] == "F":
        woman += 1
print(f"man {man}\nwoman {woman}")
````

1.5 Создайте `pd.DataFrame` `contributors`, имеющий столбцы `id`, `username` и `sex`.

````python
import pandas as pd

di = {}

for i in ["id", "username", "sex"]:
    di.update(dict.fromkeys([i], [j[i] for j in k]))
    
contributors = pd.DataFrame(di)

contributors
````

### pickle

2.1 На основе файла `contributors_sample.json` создайте словарь следующего вида: 
```
{
    должность: [список username людей, занимавших эту должность]
}
```

````python
jobs = {}
for i in k:
    for j in i["jobs"]:
        try:
            jobs[j].append(i["username"])
        except KeyError:
            t = [i["username"]]
            jobs.update({j: t})
jobs
````
